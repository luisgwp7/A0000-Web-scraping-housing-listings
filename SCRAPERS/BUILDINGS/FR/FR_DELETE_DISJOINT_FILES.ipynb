# Libraries and modules


```python
from pathlib import Path
from FR.directories import master_directory
```

# Delete disjoint files

```python
# Directories 
master_dir = master_directory()
front_directory = f'{master_dir}/FR/FRONTPAGE/JOIN' 
internal_directory = f'{master_dir}/FR/INTERNAL/JOIN'
```

## Internal disjoint files

```python
directory = Path(front_directory)
file_extensions = ['csv', 'txt', 'xlsx']

for file_path in directory.rglob('*'):
    if file_path.is_file() and file_path.suffix[1:] in file_extensions:
        file_path.unlink()
        #print(f'Deleted: {file_path}')
```

## Front disjoint files 

```python
directory = Path(internal_directory)
file_extensions = ['csv', 'txt', 'xlsx']

for file_path in directory.rglob('*'):
    if file_path.is_file() and file_path.suffix[1:] in file_extensions:
        file_path.unlink()
        #print(f'Deleted: {file_path}')
```
